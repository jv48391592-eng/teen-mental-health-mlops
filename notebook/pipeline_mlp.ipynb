```python

#=0.3=

!pip install -q wandb kagglehub pandas numpy matplotlib seaborn scikit-learn

import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
import kagglehub
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

print(" Bibliotecas instaladas com sucesso")

# ====================== CONFIGURAÇÃO REPRODUTIBILIDADE ======================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)

# ====================== LOGIN W&B ======================
API_KEY = "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"  # Substitua pela sua chave se necessário
wandb.login(key=API_KEY, relogin=True, force=True)

# ====================== 1. CARREGAMENTO E ANÁLISE EXPLORATÓRIA ======================
path = kagglehub.dataset_download("algozee/teenager-menthal-healy")
csv_path = os.path.join(path, "Teen_Mental_Health_Dataset.csv")
df_raw = pd.read_csv(csv_path)

print(f"Dataset carregado: {df_raw.shape}")
print("Colunas:", df_raw.columns.tolist())
print("\nValores ausentes por coluna:\n", df_raw.isnull().sum())
print("Duplicatas:", df_raw.duplicated().sum())

# ====================== 2. PRÉ-PROCESSAMENTO E LIMPEZA ======================
df = df_raw.copy()

# 2.1 Remoção de duplicatas
df = df.drop_duplicates()

# 2.2 Imputação de valores ausentes
num_cols = df.select_dtypes(include=np.number).columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()

# Justificativa: Mediana é robusta a outliers e distribuições assimétricas (comum em dados comportamentais)
imputer_num = SimpleImputer(strategy='median')
df[num_cols] = imputer_num.fit_transform(df[num_cols])

# Para categóricas: valor constante "missing" para preservar informação de ausência
df[cat_cols] = df[cat_cols].fillna('missing')

# 2.3 Encoding da variável alvo
le = LabelEncoder()
df['depression_label'] = le.fit_transform(df['depression_label'])
print("Classes da target:", le.classes_)

# 2.4 One-Hot Encoding nas categóricas
df = pd.get_dummies(df, columns=[col for col in cat_cols if col != 'depression_label'], drop_first=True)

print(f" Dados limpos: {df.shape}")

# ====================== 3. SELEÇÃO DE VARIÁVEIS (Feature Selection) ======================
X = df.drop(columns=['depression_label'])
y = df['depression_label']

# 3.1 Correlação de Pearson
corr = pd.Series({col: abs(X[col].corr(y)) for col in X.select_dtypes(include=np.number).columns}, name='pearson')

# 3.2 Mutual Information
mi = mutual_info_classif(X.select_dtypes(include=np.number), y, random_state=42)
mi = pd.Series(mi, index=X.select_dtypes(include=np.number).columns, name='mutual_info')

# 3.3 Importância via Random Forest
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X.select_dtypes(include=np.number), y)
rf_imp = pd.Series(rf.feature_importances_, index=X.select_dtypes(include=np.number).columns, name='rf_importance')

# Ranking final
feature_scores = pd.concat([corr, mi, rf_imp], axis=1).fillna(0)
feature_scores = feature_scores / feature_scores.max()
feature_scores['final_score'] = feature_scores.mean(axis=1)
feature_scores = feature_scores.sort_values('final_score', ascending=False)

print("\n Ranking de Features:")
display(feature_scores)

# Seleção final (threshold > 0.3)
selected_features = feature_scores[feature_scores['final_score'] > 0.3].index.tolist()
print(f"\n Features selecionadas ({len(selected_features)}): {selected_features}")

X = X[selected_features]

# ====================== 4. DIVISÃO TREINO/TESTE + ESCALONAMENTO ======================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ====================== 5. DATASET E DATALOADER ======================
class TeenDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y.values if isinstance(y, pd.Series) else y, dtype=torch.long)

    def __len__(self): return len(self.y)
    def __getitem__(self, idx): return self.X[idx], self.y[idx]

train_ds = TeenDataset(X_train_scaled, y_train)
test_ds = TeenDataset(X_test_scaled, y_test)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=64, shuffle=False)

# ====================== 6. ARQUITETURA MLP (com possibilidade de experimentação) ======================
class MLP(nn.Module):
    def __init__(self, input_dim, hidden1=128, hidden2=64, dropout=0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden1),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden1, hidden2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden2, 2)  # 2 classes
        )

    def forward(self, x):
        return self.net(x)

# ====================== 7. FUNÇÃO DE TREINAMENTO ======================
def train_model(config_name="baseline", hidden1=128, hidden2=64, dropout=0.3, lr=0.001, epochs=50):
    wandb.init(project="mlops-teen-mental-health",
               name=f"mlp-{config_name}",
               config={
                   "architecture": "MLP",
                   "layers": [X_train_scaled.shape[1], hidden1, hidden2, 2],
                   "activation": "ReLU",
                   "dropout": dropout,
                   "optimizer": "Adam",
                   "learning_rate": lr,
                   "epochs": epochs,
                   "batch_size": 64,
                   "patience": 10
               })

    model = MLP(input_dim=X_train_scaled.shape[1], hidden1=hidden1, hidden2=hidden2, dropout=dropout)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    best_loss = float('inf')
    patience = 10
    counter = 0
    train_losses, val_losses = [], []

    for epoch in range(epochs):
        # Treinamento
        model.train()
        train_loss = 0
        for bx, by in train_loader:
            optimizer.zero_grad()
            outputs = model(bx)
            loss = criterion(outputs, by)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        train_loss /= len(train_loader)
        train_losses.append(train_loss)

        # Validação
        model.eval()
        val_loss = 0
        correct, total = 0, 0
        with torch.no_grad():
            for bx, by in test_loader:
                outputs = model(bx)
                loss = criterion(outputs, by)
                val_loss += loss.item()
                _, pred = torch.max(outputs, 1)
                total += by.size(0)
                correct += (pred == by).sum().item()
        val_loss /= len(test_loader)
        acc = 100 * correct / total
        val_losses.append(val_loss)

        wandb.log({"train_loss": train_loss, "val_loss": val_loss, "accuracy": acc, "epoch": epoch+1})

        print(f"Epoch {epoch+1:2d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Acc: {acc:.2f}%")

        # Early Stopping
        if val_loss < best_loss:
            best_loss = val_loss
            counter = 0
            torch.save(model.state_dict(), f"best_model_{config_name}.pth")
        else:
            counter += 1
            if counter >= patience:
                print("Early stopping!")
                break

    # Plot da curva de loss
    plt.figure(figsize=(10,5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title(f'Curva de Loss - MLP ({config_name})')
    plt.xlabel('Época')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.savefig(f"loss_curve_{config_name}.png")
    wandb.log({"loss_curve": wandb.Image(f"loss_curve_{config_name}.png")})

    # Avaliação final
    model.load_state_dict(torch.load(f"best_model_{config_name}.pth"))
    model.eval()

    all_preds, all_labels = [], []
    with torch.no_grad():
        for bx, by in test_loader:
            outputs = model(bx)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.numpy())
            all_labels.extend(by.numpy())

    print(f"\n=== RELATÓRIO DE CLASSIFICAÇÃO - {config_name.upper()} ===")
    print(classification_report(all_labels, all_preds, target_names=["No Depression", "Depression"]))

    # Matriz de Confusão
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=["No Dep", "Dep"], yticklabels=["No Dep", "Dep"])
    plt.title(f'Matriz de Confusão - {config_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig(f"confusion_matrix_{config_name}.png")
    wandb.log({"confusion_matrix": wandb.Image(f"confusion_matrix_{config_name}.png")})

    wandb.finish()
    return model

# ====================== 8. EXPERIMENTOS (obrigatório para justificativa) ======================

print("Iniciando experimentos...")

# Experimento 1: Baseline (configuração atual)
model1 = train_model(config_name="baseline", hidden1=128, hidden2=64, dropout=0.3, lr=0.001)

# Experimento 2: Rede mais larga
model2 = train_model(config_name="wider", hidden1=256, hidden2=128, dropout=0.3, lr=0.001)

# Experimento 3: Menor dropout + lr ligeiramente menor
model3 = train_model(config_name="low_dropout", hidden1=128, hidden2=64, dropout=0.2, lr=0.0005)

print("\n Todos os experimentos concluídos e versionados no Weights & Biases!")

# ====================== SALVAR ARTIFACTS FINAIS ======================
run = wandb.init(project="mlops-teen-mental-health", name="final-artifacts")
artifact = wandb.Artifact("teen_mental_health_clean", type="dataset")
artifact.add_file("clean_teen_mental_health.csv") if 'clean_teen_mental_health.csv' in os.listdir() else None
wandb.log_artifact(artifact)
wandb.finish()

print("\n Pipeline completo finalizado")
```
