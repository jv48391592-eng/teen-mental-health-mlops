```python
# ==================== CELULA 3: DOWNLOAD + RAW ARTIFACT ====================
path = kagglehub.dataset_download("algozee/teenager-menthal-healy")
print("Path do dataset:", path)

csv_path = os.path.join(path, "Teen_Mental_Health_Dataset.csv")
df_raw = pd.read_csv(csv_path)

print(f"Dataset carregado: {df_raw.shape}")
print("Colunas:", df_raw.columns.tolist())
display(df_raw.head())

# Log raw
run = wandb.init(project="mlops-teen-mental-health", name="raw-data-upload", job_type="data-upload")
artifact = wandb.Artifact("teen_mental_health_raw", type="dataset")
artifact.add_file(csv_path)
wandb.log_artifact(artifact)
wandb.finish()
print("✅")
```

```python
# ==================== CELULA 4: LIMPEZA ====================
df = df_raw.copy()

# Remover duplicatas
df = df.drop_duplicates()

# Missing values
df = df.fillna(df.median(numeric_only=True))
df = df.fillna('missing')

# Target binária
le = LabelEncoder()
df['depression_label'] = le.fit_transform(df['depression_label'])
print("Classes:", le.classes_)

# Log clean
run = wandb.init(project="mlops-teen-mental-health", name="clean-data", job_type="data-cleaning")
clean_path = "clean_teen_mental_health.csv"
df.to_csv(clean_path, index=False)
artifact = wandb.Artifact("teen_mental_health_clean", type="dataset")
artifact.add_file(clean_path)
wandb.log_artifact(artifact)
wandb.finish()
print("✅")
```

```python
# ==================== CELULA 5: FEATURE SELECTION + SPLIT ====================
# Features numéricas/categóricas (ajuste se necessário)
feature_cols = [col for col in df.columns if col != 'depression_label']

X = df[feature_cols]
y = df['depression_label']

# Seleção rápida (RF + Mutual Info)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X.select_dtypes(include=[np.number]), y)

# Split estratificado
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Log split
run = wandb.init(project="mlops-teen-mental-health", name="train-test-split", job_type="split")
wandb.log({"train_size": len(X_train), "test_size": len(X_test)})
wandb.finish()
```

```python
# ==================== CELULA 6: PYTORCH DATASET ====================
numeric_cols = X_train.select_dtypes(include=[np.number]).columns
X_train_num = X_train[numeric_cols]
X_test_num = X_test[numeric_cols]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_num)
X_test_scaled = scaler.transform(X_test_num)

class TeenDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y.values, dtype=torch.long)
    def __len__(self): return len(self.y)
    def __getitem__(self, idx): return self.X[idx], self.y[idx]

train_ds = TeenDataset(X_train_scaled, y_train)
test_ds = TeenDataset(X_test_scaled, y_test)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=64, shuffle=False)
```

```python
# ==================== CELULA 7: MLP + TRAINING ====================
class MLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 2)
        )
    def forward(self, x):
        return self.net(x)

model = MLP(input_dim=X_train_scaled.shape[1])
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

config = {"epochs": 50, "batch_size": 64, "lr": 0.001, "dropout": 0.3, "architecture": "MLP-128-64"}

run = wandb.init(project="mlops-teen-mental-health", name="mlp-training-final", config=config)

best_loss = float('inf')
patience = 10
counter = 0

for epoch in range(config["epochs"]):
    model.train()
    train_loss = 0
    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    model.eval()
    val_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for batch_x, batch_y in test_loader:
            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)
            val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += batch_y.size(0)
            correct += (predicted == batch_y).sum().item()

    avg_train = train_loss / len(train_loader)
    avg_val = val_loss / len(test_loader)
    acc = 100 * correct / total

    wandb.log({"train_loss": avg_train, "val_loss": avg_val, "accuracy": acc, "epoch": epoch})
    print(f"Epoch {epoch+1:2d} | Train Loss: {avg_train:.4f} | Val Loss: {avg_val:.4f} | Acc: {acc:.2f}%")

    if avg_val < best_loss:
        best_loss = avg_val
        counter = 0
        torch.save(model.state_dict(), "best_model.pth")
    else:
        counter += 1
        if counter >= patience:
            print("Early stopping!")
            break

artifact = wandb.Artifact("best_mlp_model", type="model")
artifact.add_file("best_model.pth")
wandb.log_artifact(artifact)
wandb.finish()
print("✅")
```

```python
# ==================== CELULA 8: AVALIAÇÃO FINAL + ANÁLISE CRÍTICA ====================
model.load_state_dict(torch.load("best_model.pth"))
model.eval()

all_preds = []
all_labels = []
with torch.no_grad():
    for batch_x, batch_y in test_loader:
        outputs = model(batch_x)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.numpy())
        all_labels.extend(batch_y.numpy())

from sklearn.metrics import classification_report, confusion_matrix

# CORREÇÃO: converte classes para string
target_names = [str(c) for c in le.classes_]

print(classification_report(all_labels, all_preds, target_names=target_names))

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Best MLP')
plt.show()

# Log no W&B
run = wandb.init(project="mlops-teen-mental-health", name="final-evaluation")
wandb.log({"confusion_matrix": wandb.Image(plt)})
wandb.finish()
```
